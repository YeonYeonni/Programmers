```python
def solution(numbers, hand):
    answer = ''
    
    # 왼손, 오른손 초기화
    x, y = '*', '#'
    
    # 거리 계산용
    dx, dy = 0, 0
    
    # *, # 처리용
    star = {'* - 0':1,
            '# - 0':1,
            '* - 8':2,
            '# - 8':2,
            '* - 5':3,
            '# - 8':3,
            '* - 2':4,
            '# - 2':4}
    
    for i in numbers:
        
        # 왼손
        if i in [1, 4, 7]:
            answer += 'L'
            x = i
            
        # 오른손
        elif i in [3, 6, 9]:
            answer += 'R'
            y = i
            
        # 가운데
        else:
            # 거리를 계산
            if type(x) == int:
                dx = abs(x - i)
                
            if type(x) == int:
                dy = abs(y - i)
            
            # 다음 숫자가 왼손과 더 가까울 경우
            if dx < dy:
                answer += 'L'
                x = i
            
            # 다음 숫자가 오른손과 더 가까울 경우
            elif dx > dy:
                answer += 'R'
                y = i
                
            # 다음 숫자의 거리가 같은 경우
            elif dx == dy:
                
                # 오른손 잡이
                if hand == 'right':
                    answer += 'R'
                    y = i
                    
                # 왼손 잡이
                else:
                    answer += 'L'
                    x = i
    
    return answer
```
